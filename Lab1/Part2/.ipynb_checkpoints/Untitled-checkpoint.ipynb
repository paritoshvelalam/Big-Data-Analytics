```r
fludata2 <- fludata[,c(1,2,3,7,4,5,6)]
options(repr.plot.width=20, repr.plot.height=10)
library(ggplot2)
p2 <- ggplot(fludata2, aes(x = as.character(Week))) 
p2 <- p2 + theme(panel.background = element_rect(fill = "white", colour = "black"),axis.text.x=element_text(angle = 45, hjust = 1,size=14, face="bold"),axis.title.x=element_text(size=14, face="bold"),axis.text.y=element_text(size=14, face="bold"),axis.title.y=element_text(size=14, face="bold"))
p2 <- p2 + geom_bar(aes(y=Total.A, fill= "A"), stat="identity", color="black")
p2 <- p2 + geom_bar(aes(y=Total.B,fill = "B"), stat="identity", color="black")
p2 <- p2 + geom_line(aes(y=X..Positive*390.2288, group = "Percent Positive", colour="Percent Positive", linetype="Percent Positive"), stat="identity", size =2)
p2 <- p2 + geom_line(aes(y=Percent.Positive.A*390.2288, group = "% Positive Flu A", colour="% Positive Flu A", linetype="% Positive Flu A"), stat="identity", size=2)
p2 <- p2 + geom_line(aes(y=Percent.Positive.B*390.2288, group = "% Positive Flu B",colour="% Positive Flu B", linetype="% Positive Flu B"), stat="identity", size=2)
p2 <- p2 + scale_y_continuous(breaks = seq(0, 10000, by=1000), limits=c(0,10000),expand = c(0, 0),sec.axis = sec_axis(~ . /390.2288, name = "Percent Positive"))
p2 <- p2 + labs(title="Influenza Positive Tests Reported to CDC by U.S. Clinical Laboratories, National Laboratories, National Summary, 2018-19 Season",x= "Week", y="Number of Positive Specimens",fill="")
p2 <- p2 + theme(axis.title.y.right = element_text(angle = 90))
p2 <- p2 + scale_colour_manual(values=c("Percent Positive"="black","% Positive Flu A"="darkgoldenrod2","% Positive Flu B"="green"),name="")
p2 <- p2 + scale_linetype_manual(values=c("Percent Positive"="solid","% Positive Flu A"="dashed","% Positive Flu B"="dotted"),name="")
p2 <- p2 + scale_fill_manual(values=c("yellow","green4"))
p2 <- p2 + guides(fill = guide_legend(order = 1))
p2 <- p2 + theme(plot.title = element_text(size=18, face="bold"),legend.key = element_rect(colour = "transparent", fill = "white"),legend.justification = c(1,1), legend.position = c(0.8,0.8), legend.text = element_text(size = 18),legend.key.width = unit(3,"cm"))
p2
```
