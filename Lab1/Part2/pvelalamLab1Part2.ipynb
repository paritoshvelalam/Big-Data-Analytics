```r
library("ggmap")
library("maptools")
library("maps")
library("ggplot2")
data <- read.csv("heatmapdata404.csv")
data$levels = sapply(data$ACTIVITY.LEVEL, function(x) gsub("Level ", "", as.character(x)))
data$levels = as.numeric(data$levels)
state = c()
state_level = c()
for(i in unique(data$STATENAME)){
    statelevels = data[which(data$STATENAME == i),]
    state = append(state, i)
    state_level = append(state_level, max(statelevels$levels))
}

statewithlevels = data.frame(state, state_level)
statemap = map_data("state")

region_level = c()
for(i in 1:nrow(statemap)){
    region = statemap$region[i]
    index = grep(region, tolower(statewithlevels$state))
    region_level = append(region_level, statewithlevels$state_level[index[1]])
}

statemap$region_level = region_level

statemap$final_regionlevel = sapply(statemap$region_level, function(x) ifelse(x %in% c(0,2), "Minimal", ifelse(x %in% c(4,6), "Low", ifelse(x %in% c(7,8), "Moderate", "High"))))

CDC <- ggplot(statemap, aes(long, lat)) + geom_polygon(aes(group = group, fill = final_regionlevel),color = "black") 
CDC <- CDC + scale_fill_manual(values = c("red","yellow","green","orange"),name="Flu Level")+coord_quickmap() 
CDC <- CDC + labs(title="CDC Flu Map") + theme_void()+theme(legend.position = c(0.8,0.9),plot.title = element_text(size=24, face="bold")) 
CDC <- CDC + theme(legend.text = element_text(size = 18),legend.title = element_text(size = 22))
CDC
```

```r
library("ggplot2")
library("reshape2")
options(repr.plot.width=15, repr.plot.height=10) # Plot dimensions
plotdata <- read.csv("newyorkplot.csv",header =T, skip=1)
plotdata <- plotdata[14:65,4:10]
plotdata <- melt(plotdata,id=c("WEEK"))
p8 <- ggplot()
p8 <- p8 +  geom_bar(data=subset(plotdata,variable %in% c("TOTAL.A","TOTAL.B")),aes(x=factor(WEEK),y=value,fill=variable),stat="identity")
p8 <- p8 +  scale_fill_manual(values = c("TOTAL.A" = "yellow", "TOTAL.B" = "green4"),labels=c("A","B")) 
p8 <- p8 +  geom_line(data=subset(plotdata, variable %in% c("PERCENT.A","PERCENT.B","PERCENT.POSITIVE")),
            aes(x=factor(WEEK),y=value*38,group=variable,colour=variable,linetype=variable),size=2)
p8 <- p8 +  scale_colour_manual(values=c("PERCENT.A"="darkgoldenrod2","PERCENT.B"="green","PERCENT.POSITIVE"="black"),name="",labels=(c("Percent Positive","% Positive FluA","% Positive FluB")))
p8 <- p8 +  scale_linetype_manual(values=c("PERCENT.A"="dashed","PERCENT.B"="dotted","PERCENT.POSITIVE"="solid"),name="",labels=(c("Percent Positive","% Positive FluA","% Positive FluB"))) 
p8 <- p8 +  scale_y_continuous(breaks = seq(0, 3000, by=100), limits=c(0,3000),expand = c(0, 0),sec.axis = sec_axis(~./38, name = "Percent Positive"))
p8 <- p8 +  labs(title=paste0("Influenza Positive Tests Reported to CDC by U.S. Public Health Laboratories 2017-18,2018-2019 Season","\n","NEW York State"), x= "Week", y="Number of Positive Specimens",fill="",variable="")
p8 <- p8 +  theme(panel.background = element_rect(fill = "white", colour = "black"),axis.text.x=element_text(angle = 45, hjust = 1,size=14, face="bold"),axis.title.x=element_text(size=14, face="bold"),axis.text.y=element_text(size=14, face="bold"),axis.title.y=element_text(size=14, face="bold"))
p8 <- p8 +  theme(plot.title = element_text(size=18, face="bold"),legend.key = element_rect(colour = "transparent", fill = "white"),legend.justification = c(1,1), legend.position = c(0.8,0.8), legend.text = element_text(size = 18),legend.key.width = unit(3,"cm"))
p8 <- p8 +  theme(axis.title.y.right = element_text(angle = 90))
p8
```
